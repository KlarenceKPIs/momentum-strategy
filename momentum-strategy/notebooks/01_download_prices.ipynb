```python
import yfinance as yf
import pandas as pd

# Get current S&P 500 tickers from Wikipedia
sp500_url = 'https://en.wikipedia.org/wiki/List_of_S%26P_500_companies'
table = pd.read_html(sp500_url)
tickers = table[0]['Symbol'].tolist()

# Clean up tickers for yfinance
tickers = [t.replace('.', '-') for t in tickers]

# Set your date range
start_date = '2010-01-01'
end_date = '2024-12-31'

# Download monthly prices
def download_price_data(tickers):
    all_data = yf.download(
        tickers,
        start=start_date,
        end=end_date,
        interval='1mo',
        group_by='ticker',
        auto_adjust=True,
        threads=True,
        progress=True
    )
    return all_data

# Run and save
price_data = download_price_data(tickers)
price_data.to_pickle(f"{data_path}/sp500_monthly_prices.pkl")
```
